```python
import os
from PIL import Image
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Path asli dataset
input_base_path = r"D:\KULIAH\SEMESTER 5\PENGHLIHATAN DAN ANALISIS CITRA\DATASET\Klasifikasi\ORI"
output_base_path = r"D:\KULIAH\SEMESTER 5\PENGHLIHATAN DAN ANALISIS CITRA\DATASET\Klasifikasi\Processed"

# Ukuran target resize
target_size = (224, 224)

# Jumlah augmentasi per gambar
augmentation_per_image = 5

# Fungsi untuk membuat folder jika belum ada
def ensure_dir_exists(path):
    if not os.path.exists(path):
        os.makedirs(path)

# ImageDataGenerator untuk augmentasi
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Loop setiap label di folder input
for label in os.listdir(input_base_path):
    label_input_path = os.path.join(input_base_path, label)
    label_output_path = os.path.join(output_base_path, label)

    # Pastikan folder output ada
    ensure_dir_exists(label_output_path)

    # Loop setiap file dalam folder label
    for file_name in os.listdir(label_input_path):
        file_path = os.path.join(label_input_path, file_name)

        # Pastikan file adalah gambar
        if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            try:
                # Resize gambar
                img = Image.open(file_path).convert("RGB")
                img_resized = img.resize(target_size)
                resized_output_path = os.path.join(label_output_path, f"resized_{file_name}")
                img_resized.save(resized_output_path)

                # Augmentasi gambar
                img_array = np.array(img_resized)
                img_array = np.expand_dims(img_array, axis=0)

                i = 0
                for batch in datagen.flow(img_array, batch_size=1, save_to_dir=label_output_path,
                                          save_prefix=f"aug_{os.path.splitext(file_name)[0]}",
                                          save_format='jpg'):
                    i += 1
                    if i >= augmentation_per_image:
                        break

            except Exception as e:
                print(f"Error processing file {file_path}: {e}")

print("Resizing dan augmentasi selesai.")
```
